```python
import pandas as pd
import shutil
import os
from tqdm.notebook import tqdm
```

```python
csv_path = 'C:\\Users\\wq_ysw\\Desktop\\Lidar\\re.xlsx'
df = pd.read_excel(csv_path, sheet_name = 'Sheet1')
df = df[df['검수자']=='이인혁']
print(len(df))
df.head()
```

```python
df['파일'] = df['파일명'].apply(lambda x : x.split('.egg')[0])
df.head()
```

```python
단방향_path = 'X:\\dan'
전방향_path = 'X:\\jeon'

ssd_path_단방향 = 'V:\\dan'
ssd_path_전방향 = 'V:\\jeon'
```

```python
단방향_리스트 = os.listdir(단방향_path)
전방향_리스트 = os.listdir(전방향_path)

단방향_리스트 = sorted(단방향_리스트)
전방향_리스트 = sorted(전방향_리스트)
print(len(단방향_리스트))
print(len(전방향_리스트))
print(단방향_리스트)
print(전방향_리스트)
```

```python
df_new = pd.DataFrame(columns = ['파일','mosaic', '단방향', '전방향'])
df_new['파일'] = df['파일']
df_new['mosaic'] = False
df_new['단방향'] = False
df_new['전방향'] = False
df_new
```

```python
for i, file in enumerate(df_new['파일']):
    if file in 단방향_리스트:
        df_new.iloc[i, 2] = True
    elif file in 전방향_리스트:
        df_new.iloc[i, 3] = True
    else :
        df_new.iloc[i, 1] = True
        
df_new
```

```python
def get_dir_size(path='.'):
    total = 0
    with os.scandir(path) as it:
        for entry in it:
            if entry.is_file():
                total += entry.stat().st_size
            elif entry.is_dir():
                total += get_dir_size(entry.path)
    return total
```

### 폴더 용량 확인

```python
print(len(df_new[df_new['전방향'] == True]))

for i, file_name in enumerate(df_new[df_new['전방향'] == True]['파일']):
    total = 0
    scan_path = os.path.join(전방향_path, file_name)
    total += get_dir_size(scan_path)
print(total)
```

### 파일 옮기기

```python
for i, file_name in enumerate(tqdm(df_new[df_new['전방향'] == True].iloc[:21,0])):
    print('{}번  {}'.format(i, file_name))
    before_path = os.path.join(전방향_path, file_name)
    obj_path = os.path.join(ssd_path_전방향, file_name)
    
    try:
        shutil.copytree(before_path, obj_path)
    except FileExistsError:
        continue    
```
