```python
import os
import shutil
from tqdm.notebook import tqdm
import pandas as pd
```

```python
가공완료_단방향_경로 = 'X:\\dan'
가공완료_전방향_경로 = 'X:\\jeon'

test_path = os.path.join("C:\\Users\wq_ysw\Desktop","Lidar","2020-10-27-09-42-11-02")
```

```python
def tracking_id_check(폴더):
    try:
        label_list = sorted(os.listdir(os.path.join(폴더,'lidar','lidar_label')))
    except:
        print(폴더)
        return
    
    result = {}
    
    for i in label_list:
        with open(os.path.join(폴더,'lidar','lidar_label', i), mode ='rt', encoding = 'utf-8') as label:
            line = None
            tracking_list = []
            
            while line != '':
                line = label.readline()
                tracking_list.append(line.split(' ')[0])
            
            count_result = []
            
            for tracking_id in tracking_list:
                count_value = tracking_list.count(tracking_id)
                if count_value > 1:
                    count_result.append((tracking_id, str(count_value)+"번"))
            
            if len(count_result) > 0:
                count_result = list(set(count_result))
                # i에서 중복이 있는지 없는지 확인한 결과가 count_result
                
                
                result[i] = count_result
    if len(result) > 0 :
        print(폴더)
        
        for key, value in result.items():
            print("{}  :  {}".format(key, value))
        
        print("\n")
    return result
```

```python
df_단 = pd.DataFrame(columns = ['폴더', 'label_번호', '중복_tracking_id'])
j = 0

for 폴더 in tqdm(os.listdir(가공완료_단방향_경로)):
    result = tracking_id_check(os.path.join(가공완료_단방향_경로, 폴더))
    
    for i in range(len(result)):
        popitem = result.popitem()
        df_단.loc[j] = [폴더, popitem[0], popitem[1]]
        j = j+1
            
df_단.to_csv('C:\\Users\\wq_ysw\\Desktop\\Lidar\\tracking_id_중복_단방향.csv', mode='w', index=False, encoding='utf-8-sig')
```
